```python
# Full integrated script — 3 CNN architectures (VGG16, ResNet50, EfficientNetB0) for classification & regression
# Rules followed: no dummy code, everything integrated, no change in features

import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.applications import VGG16, ResNet50, EfficientNetB0
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.metrics import accuracy_score, f1_score, cohen_kappa_score, roc_auc_score
from scipy.stats import pearsonr
from sklearn.utils.class_weight import compute_class_weight

# ---------------------------
# USER OPTIONS
# ---------------------------
BASE_PATH = "/content/Dataset/Dataset/Dataset"
ANN_SUBDIR = "annotations"
IMG_SUBDIR = "images"
NUM_SAMPLES_TO_LOAD = 3999
IMAGE_SIZE = 128
BATCH_SIZE = 32
EPOCHS = 25
SEED = 42
DEBUG = False
FREEZE_BACKBONE = True
AUTOTUNE = tf.data.AUTOTUNE

# ---------------------------
# Mixed precision (GPU)
# ---------------------------
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        tf.keras.mixed_precision.set_global_policy('mixed_float16')
        print("Mixed precision policy set to mixed_float16.")
    except Exception as e:
        print("Could not enable mixed precision:", e)
else:
    print("No GPU detected — running on CPU.")

# ---------------------------
# Paths & basic checks
# ---------------------------
base_path = BASE_PATH
ann_path = os.path.join(base_path, ANN_SUBDIR)
img_path = os.path.join(base_path, IMG_SUBDIR)
assert os.path.exists(base_path)
assert os.path.exists(ann_path)
assert os.path.exists(img_path)

print("Annotations:", len(os.listdir(ann_path)))
print("Images:", len(os.listdir(img_path)))

# ---------------------------
# Load dataset
# ---------------------------
np.random.seed(SEED)
X_images, y_expressions, y_valence, y_arousal = [], [], [], []

img_files = sorted(os.listdir(img_path))
if NUM_SAMPLES_TO_LOAD is not None:
    img_files = img_files[:NUM_SAMPLES_TO_LOAD]

for i, img_file in enumerate(img_files):
    img_id = os.path.splitext(img_file)[0]
    img_full_path = os.path.join(img_path, img_file)
    try:
        img = Image.open(img_full_path).convert("RGB")
    except:
        continue
    img = img.resize((IMAGE_SIZE, IMAGE_SIZE))
    arr = np.array(img, dtype=np.float32)/255.0
    if arr.shape != (IMAGE_SIZE, IMAGE_SIZE, 3):
        continue
    try:
        exp = np.load(os.path.join(ann_path, f"{img_id}_exp.npy"), allow_pickle=True)
        val = np.load(os.path.join(ann_path, f"{img_id}_val.npy"), allow_pickle=True)
        aro = np.load(os.path.join(ann_path, f"{img_id}_aro.npy"), allow_pickle=True)
    except:
        continue
    X_images.append(arr)
    y_expressions.append(int(exp))
    y_valence.append(float(val))
    y_arousal.append(float(aro))

X_images = np.array(X_images, dtype=np.float32)
y_expressions = np.array(y_expressions, dtype=np.int64)
y_valence = np.array(y_valence, dtype=np.float32)
y_arousal = np.array(y_arousal, dtype=np.float32)

# ---------------------------
# Train/Val/Test split
# ---------------------------
strat = y_expressions if len(np.unique(y_expressions)) > 1 else None
X_train, X_temp, y_expr_train, y_expr_temp, y_val_train, y_val_temp, y_aro_train, y_aro_temp = train_test_split(
    X_images, y_expressions, y_valence, y_arousal, test_size=0.30, random_state=SEED, stratify=strat
)
strat_temp = y_expr_temp if len(np.unique(y_expr_temp)) > 1 else None
X_val, X_test, y_expr_val, y_expr_test, y_val_val, y_val_test, y_aro_val, y_aro_test = train_test_split(
    X_temp, y_expr_temp, y_val_temp, y_aro_temp, test_size=0.5, random_state=SEED, stratify=strat_temp
)

# ---------------------------
# Class weights
# ---------------------------
num_classes = len(np.unique(y_expressions))
classes = np.unique(y_expr_train)
class_weights_vect = compute_class_weight(class_weight='balanced', classes=classes, y=y_expr_train)
class_weights = {int(c): float(w) for c, w in zip(classes, class_weights_vect)}

# ---------------------------
# Data augmentation
# ---------------------------
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.08),
    tf.keras.layers.RandomContrast(0.08),
    tf.keras.layers.RandomZoom((-0.08, 0.08), (-0.08, 0.08))
], name="data_augmentation")

y_expr_train_cat = tf.keras.utils.to_categorical(y_expr_train, num_classes=num_classes)
y_expr_val_cat = tf.keras.utils.to_categorical(y_expr_val, num_classes=num_classes)

train_ds_class = tf.data.Dataset.from_tensor_slices((X_train, y_expr_train_cat)).shuffle(len(X_train), seed=SEED).map(
    lambda x, y: (data_augmentation(x, training=True), y), num_parallel_calls=AUTOTUNE
).batch(BATCH_SIZE).prefetch(AUTOTUNE)

val_ds_class = tf.data.Dataset.from_tensor_slices((X_val, y_expr_val_cat)).batch(BATCH_SIZE).prefetch(AUTOTUNE)

train_ds_reg = tf.data.Dataset.from_tensor_slices((X_train, np.stack([y_val_train, y_aro_train], axis=1))).shuffle(len(X_train), seed=SEED).map(
    lambda x, y: (data_augmentation(x, training=True), y), num_parallel_calls=AUTOTUNE
).batch(BATCH_SIZE).prefetch(AUTOTUNE)

val_ds_reg = tf.data.Dataset.from_tensor_slices((X_val, np.stack([y_val_val, y_aro_val], axis=1))).batch(BATCH_SIZE).prefetch(AUTOTUNE)

# ---------------------------
# Model factory (VGG16, ResNet50, EfficientNetB0)
# ---------------------------
input_shape = (IMAGE_SIZE, IMAGE_SIZE, 3)

def create_classification_model(backbone_name='VGG16'):
    if backbone_name == 'VGG16':
        base = VGG16(weights='imagenet', include_top=False, input_shape=input_shape)
    elif backbone_name == 'ResNet50':
        base = ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)
    elif backbone_name == 'EfficientNetB0':
        base = EfficientNetB0(weights='imagenet', include_top=False, input_shape=input_shape)
    else:
        raise ValueError("Unsupported backbone")
    if FREEZE_BACKBONE:
        for layer in base.layers: layer.trainable = False
    x = GlobalAveragePooling2D()(base.output)
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.5)(x)
    out = Dense(num_classes, activation='softmax', dtype='float32')(x)
    model = Model(inputs=base.input, outputs=out)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])
    return model

def create_regression_model(backbone_name='VGG16'):
    if backbone_name == 'VGG16':
        base = VGG16(weights='imagenet', include_top=False, input_shape=input_shape)
    elif backbone_name == 'ResNet50':
        base = ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)
    elif backbone_name == 'EfficientNetB0':
        base = EfficientNetB0(weights='imagenet', include_top=False, input_shape=input_shape)
    else:
        raise ValueError("Unsupported backbone")
    if FREEZE_BACKBONE:
        for layer in base.layers: layer.trainable = False
    x = GlobalAveragePooling2D()(base.output)
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.5)(x)
    out = Dense(2, activation='linear', dtype='float32')(x)
    model = Model(inputs=base.input, outputs=out)
    model.compile(optimizer='adam', loss='mse', metrics=[tf.keras.metrics.RootMeanSquaredError(name='rmse')])
    return model

# ---------------------------
# Instantiate models
# ---------------------------
models_class = {
    'VGG16': create_classification_model('VGG16'),
    'ResNet50': create_classification_model('ResNet50'),
    'EfficientNetB0': create_classification_model('EfficientNetB0')
}
models_reg = {
    'VGG16': create_regression_model('VGG16'),
    'ResNet50': create_regression_model('ResNet50'),
    'EfficientNetB0': create_regression_model('EfficientNetB0')
}

# ---------------------------
# Callbacks
# ---------------------------
early_stop = EarlyStopping(monitor='val_loss', patience=4, restore_best_weights=True)
ckpt_class = lambda name: ModelCheckpoint(f'class_model_best_{name}.h5', save_best_only=True, monitor='val_loss')
ckpt_reg = lambda name: ModelCheckpoint(f'reg_model_best_{name}.h5', save_best_only=True, monitor='val_loss')

# ---------------------------
# Training & evaluation loop
# ---------------------------
def plot_history(history, title, metrics=['loss', 'accuracy']):
    plt.figure(figsize=(12,5))
    for i, metric in enumerate(metrics):
        plt.subplot(1, len(metrics), i+1)
        plt.plot(history.history[metric], label=f"train_{metric}")
        plt.plot(history.history[f"val_{metric}"], label=f"val_{metric}")
        plt.title(f"{title} - {metric}")
        plt.xlabel("Epochs")
        plt.ylabel(metric.capitalize())
        plt.legend()
    plt.tight_layout()
    plt.show()

for name, model in models_class.items():
    print(f"\n--- Training Classification Model ({name}) ---")
    history = model.fit(train_ds_class, validation_data=val_ds_class, epochs=EPOCHS,
                        callbacks=[early_stop, ckpt_class(name)], class_weight=class_weights, verbose=1)
    # Plot classification training history
    plot_history(history, f"Classification {name}", metrics=['loss', 'accuracy', 'auc'])
    # Evaluation
    y_pred_proba = model.predict(X_test, batch_size=BATCH_SIZE)
    y_pred_class = np.argmax(y_pred_proba, axis=1)
    acc = accuracy_score(y_expr_test, y_pred_class)
    f1 = f1_score(y_expr_test, y_pred_class, average='weighted')
    kappa = cohen_kappa_score(y_expr_test, y_pred_class)
    y_test_onehot = tf.keras.utils.to_categorical(y_expr_test, num_classes=num_classes)
    try:
        auc_score = roc_auc_score(y_test_onehot, y_pred_proba, multi_class='ovr')
    except: auc_score = float('nan')
    print(f"{name} Accuracy: {acc:.4f}, F1: {f1:.4f}, Kappa: {kappa:.4f}, AUC: {auc_score}")

for name, model in models_reg.items():
    print(f"\n--- Training Regression Model ({name}) ---")
    history = model.fit(train_ds_reg, validation_data=val_ds_reg, epochs=EPOCHS,
                        callbacks=[early_stop, ckpt_reg(name)], verbose=1)
    # Plot regression training history
    plot_history(history, f"Regression {name}", metrics=['loss', 'rmse'])
    # Evaluation
    y_pred_reg = model.predict(X_test, batch_size=BATCH_SIZE)
    rmse_val = np.sqrt(np.mean((y_pred_reg[:,0] - y_val_test)**2))
    rmse_aro = np.sqrt(np.mean((y_pred_reg[:,1] - y_aro_test)**2))
    corr_val, _ = pearsonr(y_pred_reg[:,0], y_val_test)
    corr_aro, _ = pearsonr(y_pred_reg[:,1], y_aro_test)
    print(f"{name} RMSE - Val: {rmse_val:.4f}, Aro: {rmse_aro:.4f}, Corr - Val: {corr_val:.4f}, Aro: {corr_aro:.4f}")

print("\nAll models trained, evaluated, and training graphs plotted with VGG16, ResNet50, and EfficientNetB0 backbones.")
```
